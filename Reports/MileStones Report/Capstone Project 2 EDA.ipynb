```python
# Import all necessary libraries 

import numpy as np 
import os 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.utils import resample
import pandas as pd 
import statsmodels.api as sm 
import math as m
import seaborn as sns 
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import classification_report,confusion_matrix
from collections import Counter

```

```python
# Useful functions

# To preprocess data 
def preprocess_data(ss,unwanted): 
    fn=os.listdir(ss)
    data=load_ss_csv_to_df(fn,ss)

    data[['Depth','Delta_SSNS','Delta_SSPS','SSNS_Prefrac','SSNS_Prefrac_RMS','SSNS_Duringfrac','SSNS_Duringfrac_RMS','SSNS_Afterfrac_RMS','SSNS_Afterfrac','SSPS_Prefrac','SSPS_Prefrac_RMS','SSPS_Duringfrac_RMS','SSPS_Afterfrac_RMS','SSPS_Duringfrac','SSPS_Afterfrac','Total_Strain_Duringfrac','Total_Strain_Afterfrac']]=data[['Depth','Delta_SSNS','Delta_SSPS','SSNS_Prefrac','SSNS_Prefrac_RMS','SSNS_Duringfrac','SSNS_Duringfrac_RMS','SSNS_Afterfrac_RMS','SSNS_Afterfrac','SSPS_Prefrac','SSPS_Prefrac_RMS','SSPS_Duringfrac_RMS','SSPS_Afterfrac_RMS','SSPS_Duringfrac','SSPS_Afterfrac','Total_Strain_Duringfrac','Total_Strain_Afterfrac']].astype(float)

    data['FracHits']=data['FracHits'].astype(int)
    data.set_index('Depth',inplace=True)

    # Create Train and Test Data
    target=data['FracHits']

    features=data.drop(unwanted,axis=1)
    
    return data,features,target


# Function to read all LAS files into dataframe 
def load_ss_csv_to_df(fn,ss): 
    
    
    firstcolumn_list=[]
    secondcolumn_list=[]
    thirdcolumn_list=[]
    fourthcolumn_list=[]
    fifthcolumn_list=[]
    sixthcolumn_list=[]
    seventhcolumn_list=[]
    eighthcolumn_list=[]
    ninthcolumn_list=[]
    tenthcolumn_list=[]
    eleventhcolumn_list=[]
    twelvethcolumn_list=[]
    thirteencolumn_list=[]
    fourteencolumn_list=[]
    fifteencolumn_list=[]
    sixteencolumn_list=[]
    seventeencolumn_list=[]
    eighteencolumn_list=[]
    nineteencolumn_list=[]
    twentycolumn_list=[]
    twentyonecolumn_list=[]
    twentytwocolumn_list=[]

    for idx in range(len(fn)): 
    

        f=fn[idx]
        full_file_dir=os.path.join(ss,f)
        
        with open(full_file_dir, encoding='latin-1') as datafile:
             lines=datafile.readlines()[1:]
             
             for line in lines:
                 firstcolumn_list.append(line.split(',')[0])
                 secondcolumn_list.append(line.split(',')[1])
                 thirdcolumn_list.append(line.split(',')[2])
                 fourthcolumn_list.append(line.split(',')[3])
                 fifthcolumn_list.append(line.split(',')[4])
                 sixthcolumn_list.append(line.split(',')[5])
                 seventhcolumn_list.append(line.split(',')[6])
                 eighthcolumn_list.append(line.split(',')[7])
                 ninthcolumn_list.append(line.split(',')[8])
                 tenthcolumn_list.append(line.split(',')[9])
                 eleventhcolumn_list.append(line.split(',')[10])
                 twelvethcolumn_list.append(line.split(',')[11])
                 thirteencolumn_list.append(line.split(',')[12])
                 fourteencolumn_list.append(line.split(',')[13])
                 fifteencolumn_list.append(line.split(',')[14])
                 sixteencolumn_list.append(line.split(',')[15])
                 seventeencolumn_list.append(line.split(',')[16])
                 eighteencolumn_list.append(line.split(',')[17])
#                 nineteencolumn_list.append(line.split(',')[18])
#                 twentycolumn_list.append(line.split(',')[19])
#                 twentyonecolumn_list.append(line.split(',')[20])
#                 twentytwocolumn_list.append(line.split(',')[21])


        
        
          
    ss_dict={
         'Depth': firstcolumn_list,'SSNS_Prefrac':secondcolumn_list,
         'SSNS_Prefrac_RMS': thirdcolumn_list, 'SSNS_Duringfrac':fourthcolumn_list, 'SSNS_Duringfrac_RMS':fifthcolumn_list,
         'SSNS_Afterfrac_RMS':sixthcolumn_list, 'SSNS_Afterfrac':seventhcolumn_list, 'SSPS_Prefrac':eighthcolumn_list,
         'SSPS_Prefrac_RMS':ninthcolumn_list,'SSPS_Duringfrac_RMS':tenthcolumn_list, 'SSPS_Afterfrac_RMS':eleventhcolumn_list,
         'SSPS_Duringfrac': twelvethcolumn_list, 'SSPS_Afterfrac':thirteencolumn_list,
#         'SSNS_Prefrac_fft':fourteencolumn_list,
#         'SSNS_Duringfrac_fft':fifteencolumn_list, 'SSNS_Afterfrac_fft':sixteencolumn_list,'SSPS_Prefrac_fft':seventeencolumn_list,
#         'SSPS_Duringfrac_fft':eighteencolumn_list, 'SSPS_Afterfrac_fft':nineteencolumn_list,
          'Total_Strain_Duringfrac':fourteencolumn_list,
          'Total_Strain_Afterfrac':fifteencolumn_list,
          'Delta_SSPS':sixteencolumn_list,
          'Delta_SSNS':seventeencolumn_list,
          'FracHits':eighteencolumn_list
          
          # 'FracHits':sixteencolumn_list
            }

    data=pd.DataFrame()
    data=pd.concat([pd.Series(v,name=k) for k, v in ss_dict.items()],axis=1)

    return data 
```
